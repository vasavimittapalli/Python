```python
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

import random
words=['life','success','journey','destination']
word_choice=random.choice(words)
print('word_choice:',word_choice)
blank=[]
for letter in word_choice:
  blank.append('_')
end_of_game=False
lives=6
while not end_of_game:
  guess=input('Enter a letter to guess:').lower()
  for position in range(len(word_choice)):
     letter=word_choice[position]
     if guess in letter:
         blank[position]=letter

  if guess not in word_choice:
    lives -= 1
    if lives==0:
      end_of_game=True
      print('You lose')
  print(blank)
  if '_' not in blank:
    end_of_game=True
    print('You win')
  print(stages[lives])
```
